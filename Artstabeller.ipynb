```python
# FA bruker en annen inndeling enn RL på Artsgruppe
# Basert på at både RL og FA bruker taksonomisk sti, så kan man bruke den som et oppslagsliste for å mappe verdier fra RL.artsgruppe til FA_l
# Subset RL with unique values of columns "Taksonomisk sti" and "Artsgruppe"
AG = RL[['Taksonomisk sti', 'Artsgruppe']].drop_duplicates()
AG_2 = FA_l[['Taksonomisk sti', 'Artsgruppe']].drop_duplicates()

# Function to find corresponding Artsgruppe in AG/Rødlista based on Taksonomisk sti
def find_artsgruppe(taksonomisk_sti):
    for index, row in AG.iterrows():
        if taksonomisk_sti in row['Taksonomisk sti']:
            return row['Artsgruppe']
    return None

# Add new column "Artsgruppe_2" to FA_l
FA_l['Artsgruppe'] = FA_l['Artsgruppe'].apply(lambda x: find_artsgruppe(x) if pd.notna(x) else None)
```

# Verdsatte arter

```python
import numpy as np
import pandas as pd
```

```python
ADB = ADB.rename(columns={'Kategori 2021':"RL_status","Vitenskapelig navn id":"vitenskapelig_navn_id","Vitenskapelig navn":"vitenskapelig_navn"})
VitID2 = pd.DataFrame(data = {"vitenskapelig_navn_id":np.concatenate([forvaltningsstatus["vitenskapelig_navn_id"],ADB["vitenskapelig_navn_id"]])})
VitID = pd.DataFrame(data = {"vitenskapelig_navn_id":VitID2["vitenskapelig_navn_id"].unique()})
```

```python
AV = VitID.merge(right = ADB, how = "left", on ="vitenskapelig_navn_id")
AV = AV.merge(right = forvaltningsstatus.drop(columns=['vitenskapelig_navn']), how = "left", on = ["vitenskapelig_navn_id"])
```

```python
# Hardkoder at artene i spesielle økologiske former er fugler
AV.loc[AV["spesielle_oekologiske_former"]==1,"Artsgruppe"]="Fugler"
# Tilsvarende for prioriterte arter
AV.loc[AV["vitenskapelig_navn_id"]==3768,"Artsgruppe"]="Fugler"
AV[(AV["prioritert"]==1)]
# Og fikser deres norske navn
AV.loc[(AV["Populærnavn"].isna()) & (~AV["norsk_navn"].isna()), "Populærnavn"] = AV["norsk_navn"]
```

```python
AV.loc[AV["vitenskapelig_navn_id"]==3768]
```

```python
# "Øvrige dyr"
AV.loc[AV["Artsgruppe"].isna(), "Artsgruppe"]  = "Øvrige dyr"
```

```python
# Function to map values
# Make sure that the prioritized values occur first, so that lower values are not used when a species can have several
def map_values(row):
    #print(row)
    #if row['freda'] == 1 or row['prioritert'] == 1:
        #return 'Svært stor verdi'
    if (row['RL_status'] in ['CR', 'EN']) or (row['freda'] == 1 or row['prioritert'] == 1):
        return 'Svært stor verdi'
    elif (row['RL_status'] == 'VU') or (row["Hensynskrevende"]=="Spesielt"):
        return 'Stor verdi'
    elif row['RL_status'] in ['NT', 'DD']:
        return 'Middels verdi'
    elif (row['RL_status'] == 'LC') or (row["Hensynskrevende"]=="Annet"):
        return 'Noe verdi'    
    else:
        return "-"

# Apply the function to the DataFrame
AV['Forvaltningsverdi'] = AV.apply(map_values, axis=1)
```

```python
AV[AV['prioritert']==1][["vitenskapelig_navn","Populærnavn","RL_status","prioritert","Forvaltningsverdi"]]
```

```python
# Function to set criteria based on RL_status
def set_criteria(row):
    if row['RL_status'] == 'NT':
        row['Kriterium_NT'] = 'X'
    elif row['RL_status'] == 'DD':
        row['Kriterium_DD'] = 'X'
    elif row['RL_status'] == 'VU':
        row['Kriterium_sårbar art'] = 'X'
    elif row['RL_status'] == 'CR':
        row['Kriterium_kritisk trua art'] = 'X'
    elif row['RL_status'] == 'EN':
        row['Kriterium_sterkt trua art'] = 'X'
    elif row["Hensynskrevende"]=="Spesielt":
        row['Kriterium_Andre spesielt_hensynskrevende_arter']='X'
    elif row["Hensynskrevende"]=="Annet":
        row['Kriterium_Hensynskrevende_arter']='X'        
    return row

# Apply the function to each row
AV = AV.apply(set_criteria, axis=1)
```

```python
AV = AV.rename(columns={"RL_status":"Rødlistestatus",'prioritert':'Kriterium_Prioriterte_arter','freda':"Kriterium_Fredete_arter","spesielle_oekologiske_former":"Kriterium_Spesielle_okologiske_former"})
```

```python
# for arter som ikke har noen vurdering, slik som priorirterte (under)arter
AV.loc[AV["Vurderingsområde"].isna(),"Vurderingsområde"]="Norge" 
# Identifisere ansvarsarter
AV.loc[AV["Andel av europeisk populasjon i Norge"].isna(),"Andel av europeisk populasjon i Norge"]="-"
AV.loc[AV["Andel av europeisk populasjon i Norge"].str.contains("50"), "Kriterium_Ansvarsart"] = "1"
# Remove potential 0's,
# replace potential X's with 1's
AV.loc[:, AV.columns.str.startswith('Kriterium_')] = AV.loc[:, AV.columns.str.startswith('Kriterium_')].replace({0: np.nan, 'X': 1})
```

```python
# Snake_case it all
AV.columns = (AV.columns
                .str.replace(' ', '_', regex=True)
                .str.lower()
             )
AV.columns
```
